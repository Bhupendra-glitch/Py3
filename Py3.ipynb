```python
def sum_squares_abs_diff_squares(a: int, b: int) -> tuple:
    '''Return the sum of squares and the absolute difference of squares.

    Args:
        a, b : int - input numbers

    Returns:
        tuple - (sum of squares, absolute difference of squares)
    '''

    sum_of_squares = a**2 + b**2
    abs_diff_of_squares = abs(a**2 - b**2)
    return(sum_of_squares, abs_diff_of_squares)
```

```python
def is_odd_length_palindrome(s: str) -> bool:
    '''Check if a string is a palindrome with odd length.

    Args:
        s : str - input string

    Returns:
        bool - True if s is a palindrome with odd length, False otherwise.
    '''

    #Checking if the string has odd length

    if len(s) % 2 == 0:
      return False

    #Checking the string is palindram

    return s == s[::-1]
```

```python
def get_roll_nos(data: list, criteria=None):
    '''
    Filter roll numbers based on criteria.

    Args:
        data (list): List of tuples with roll number and marks.
        criteria (str, optional): The criteria for filtering.

    Returns:
        list: List of roll numbers matching the criteria or None if invalid criteria.
    '''

    # Calculate average marks
    avg = sum(marks for _, marks in data) / len(data)

    if criteria == 'above_average':
        return [roll for roll, marks in data if marks >= avg]
    elif criteria == 'below_average':
        return [roll for roll, marks in data if marks < avg]
    elif criteria == 'fail':
        return [roll for roll, marks in data if marks < 40]
    elif criteria == 'toppers':
        return [roll for roll, marks in data if marks >= 90]
    elif criteria is None:
        return [roll for roll, _ in data]
    else:
        return None
```

```python
#Check if the Given String is Within Quotes and Has Quotes Inside
```

```python
def within_and_has_double_quotes(s: str) -> bool:
    '''Check if the string is enclosed with double quotes and has double quotes inside.

    Args:
        s : str - input string

    Returns:
        bool - True if the string starts and ends with double quotes and has double quotes inside
    '''
    # Check if string starts and ends with double quotes
    if len(s) >= 2 and s[0] == '"' and s[-1] == '"':
        # Check if there is another double quote inside (excluding first and last)
        inner = s[1:-1]
        if '"' in inner:
            return True
    return False
```

```python
def replace_middle_with_n_times_middle(t: tuple, n: int) -> tuple:
    '''
    Replace the middle element of a tuple with `n` copies of the middle element.

    Args:
        t (tuple): A tuple with an odd number of elements.
        n (int): The number of times the middle element should be repeated.

    Returns:
        tuple: A new tuple with the middle element replaced by `n` copies.
    '''

    mid_index = leg(t) // 2  # find the middle index
    middle_element = t[mid_index]   # Get a middle element
    return t[:mid_index] + (middle_element,) * n + t[mid_index + 1:]
```

```python
def filter_students(data: dict, criteria: str) -> set:
    '''
    Takes a dictionary where keys are student names and values are lists of scores,
    filters names based on the given criteria:
     - "excellent" - average score >= 85.
     - "good" - average score >= 50 and < 85.
     - "all_pass" - all scores >= 50.
     - "balanced" - difference between min and max is <= 10.
    '''

    result = set()

    for name, scores in data.items():
        avg = sum(scores) / len(scores)

        if criteria == "excellent" and avg >= 85:
            result.add(name)
        elif criteria == "good" and 50 <= avg < 85:
            result.add(name)
        elif criteria == "all_pass" and all(s >= 50 for s in scores):
            result.add(name)
        elif criteria == "balanced" and (max(scores) - min(scores)) <= 10:
            result.add(name)

    return result
```

```python
def is_rotation(str1: str, str2: str) -> bool:
    """
    Checks if one string is a rotation of another.

    Example:
      >>>is_rotation('abcde','cdeab')
      True
      >>>is_rotation('hello','ehllo')
      False
      >>>is_rotation('hello','helol')
      False

    Args:
        str1 (str): The original string.
        str2 (str): The string to check if it's a rotation of str1.

    Returns:
        bool: True if str2 is a rotation of str1, False otherwise.
    """
    # Both strings must be of same length and non-empty
    if len(str1) != len(str2):
        return False

    # A string is a rotation of another if it's a substring of str1 + str1
    return str2 in (str1 + str1)
```

```python
def find_missing_number(nums: list, n: int) -> int:
    """Finds the missing number in a list of integers ranging from 1 to n.

    Args:
        nums (list): A list of integers from 1 to n with one number missing.
        n (int): The number n describing the range.
    Returns:
        int: The missing number in the list.
    """
    expected_sum = n * (n + 1) // 2   # Sum of 1 to n
    actual_sum = sum(set(nums))        # Sum of unique numbers in the list
    return expected_sum - actual_sum
```

```python
def merge_dictionaries(d1: dict, d2: dict):
    """
    Merges two dictionaries by summing values of common keys.

    If a key exists in both `d1` and `d2`, their values are summed.
    Keys that are unique to one dictionary are included as-is in the result.

    Args:
        d1 (dict): The first dictionary, where values are integers or floats.
        d2 (dict): The second dictionary, where values are integers or floats.

    Returns:
        dict: A new dictionary containing all keys from both `d1` and `d2`.
              Values for overlapping keys are summed, while unique keys retain their original values.
    """
    merged = d1.copy()  # start with d1
    for key, value in d2.items():
        merged[key] = merged.get(key, 0) + value  # sum if exists, else add new
    return merged
```

```python
def divide_into_almost_equal_parts(n: int, k: int) -> list:
    # Base value (equal division)
    base = n // k
    # Remainder (extra ones to distribute)
    remainder = n % k

    # Start with base values
    parts = [base] * k

    # Distribute the remainder — one extra to the first 'remainder' elements
    for i in range(remainder):
        parts[i] += 1

    return parts
```

```python
def overall_run_stats(batsman_data: dict) -> dict:
    """Returns a dict with overall run statistics with keys 'min', 'max', 'total' and 'average'."""
    all_runs = [run for runs in batsman_data.values() for run in runs]
    return {
        "min": min(all_runs),
        "max": max(all_runs),
        "total": sum(all_runs),
        "average": round(sum(all_runs) / len(all_runs))
    }


def century_rate(runs: list) -> int:
    """Returns the century rate from the given runs."""
    if not runs:
        return 0
    centuries = sum(1 for r in runs if r >= 100)
    return round((centuries / len(runs)) * 100)


def average_yearly_century_rate(batsman_data: dict) -> int:
    """Average yearly century rate over all years."""
    rates = [century_rate(runs) for runs in batsman_data.values()]
    return round(sum(rates) / len(rates))


def years_with_more_than_average_yearly_century_rate(batsman_data: dict) -> set:
    """Returns a set of years with century rate more than the average yearly century rate."""
    avg_rate = average_yearly_century_rate(batsman_data)
    return {year for year, runs in batsman_data.items() if century_rate(runs) > avg_rate}


def year_with_most_average_runs(batsman_data: dict) -> int:
    """Returns the year with the most average runs."""
    year_avg = {year: sum(runs) / len(runs) for year, runs in batsman_data.items()}
    max_avg = max(year_avg.values())
    # Return earliest year in case of tie
    return min(year for year, avg in year_avg.items() if avg == max_avg)
```

```python
def is_even_and_second_last_digit_is_two(num: int) -> bool:
    '''
    Returns True if:
      1. The number is even, AND
      2. The second last digit is 2.
    Otherwise, returns False.

    Examples:
    >>> is_even_and_second_last_digit_is_two(232)
    False
    >>> is_even_and_second_last_digit_is_two(120)
    True
    >>> is_even_and_second_last_digit_is_two(25)
    False
    >>> is_even_and_second_last_digit_is_two(13)
    False
    '''

    # Check even number
    if num % 2 != 0:
        return False

    # Convert to string to access digits
    num_str = str(abs(num))   # abs() handles negative numbers safely

    # Check if it has at least two digits
    if len(num_str) < 2:
        return False

    # Check if second last digit is '2'
    return num_str[-2] == '2'
```

```python
def increment_value_with_max_limit(d: dict, key: str, inc: int, limit: int):
    '''
    Increment the value for the given key in the dictionary by `inc`,
    but not beyond `limit`.

    Modify the value in the dictionary (no return).

    Args:
        d : dict - dictionary with integer values
        key : str - key to increment in the dictionary
        inc : int - increment value
        limit : int - maximum limit for the value
    '''
    # Step 1: Get current value
    current_value = d.get(key, 0)

    # Step 2: Increment it
    new_value = current_value + inc

    # Step 3: Apply max limit (capping)
    if new_value > limit:
        new_value = limit

    # Step 4: Update the dictionary
    d[key] = new_value
```

```python
def markdown_to_html_image(markdown_image: str) -> str:
    '''
    Converts a Markdown image link to an HTML image tag.
    Example: "![alt](url)" → '<img src="url" alt="alt">'
    '''
    # Step 1: Find positions of brackets and parentheses
    start_alt = markdown_image.find('[') + 1
    end_alt = markdown_image.find(']')
    start_src = markdown_image.find('(') + 1
    end_src = markdown_image.find(')')

    # Step 2: Extract alt text and image URL
    alt_text = markdown_image[start_alt:end_alt]
    image_url = markdown_image[start_src:end_src]

    # Step 3: Format to HTML image tag
    return f'<img src="{image_url}" alt="{alt_text}">'
```

```python
def rotate_even(l: list, k: int) -> list:
    '''
    Returns a list with the elements at the even indices rotated by k positions.

    Examples:
    >>> rotate_even([1,2,3,4,5,6],1)
    [5,2,1,4,3,6]
    >>> rotate_even([1,2,3,4,5,6,7,8],3)
    [3,2,5,4,7,6,1,8]
    '''

    # Extract even-index elements
    even_elements = l[::2]

    # Handle k greater than length
    k = k % len(even_elements)

    # Rotate right by k
    rotated = even_elements[-k:] + even_elements[:-k]

    # Replace even indices
    l[::2] = rotated

    return l
```

```python
def create_slug(text: str) -> str:
    '''
    Converts a given string into a URL-friendly slug.

    Examples:
    >>> create_slug("Hello World!")
    'hello-world'
    >>> create_slug("Python Programming 101")
    'python-programming-101'
    >>> create_slug("Data Science & AI")
    'data-science-ai'

    Args:
        text (str): The input string

    Returns:
        str: A slugified version of the string
    '''
    import re
    # Convert to lowercase
    text = text.lower()
    # Remove non-alphanumeric characters (except spaces)
    text = re.sub(r'[^a-z0-9\s-]', '', text)
    # Replace multiple spaces or hyphens with a single hyphen
    text = re.sub(r'[\s-]+', '-', text)
    # Strip hyphens from start and end
    text = text.strip('-')
    return text
```

```python
def replace_vowels_with_next_alphabet(s: str) -> str:
    """
    Replaces all vowels in the input string with the next character in the alphabet.

    Args:
        s (str): The input string to process.

    Returns:
        str: A string with all vowels replaced by their next alphabetical character.

    Note:
        This function is case-sensitive.

    Examples:
        >>> replace_vowels_with_next_alphabet("hello")
        'hflmp'
        >>> replace_vowels_with_next_alphabet("HELLO")
        'HFLMP'
    """
    vowels = "aeiouAEIOU"
    result = ""

    for ch in s:
        if ch in vowels:
            result += chr(ord(ch) + 1)
        else:
            result += ch

    return result
```

```python
def process_sentence(sentence, task):
    words = sentence.split()

    if task == "count_words":
        return len(words)

    elif task == "count_palindromes" or task == "count_palindrome_words":
        return sum(1 for w in words if w == w[::-1])

    elif task == "count_words_with_repeated_chars":
        def has_repeated_chars(word):
            return len(set(word)) != len(word)
        return sum(1 for w in words if has_repeated_chars(w))

    elif task == "words_with_max_len":
        max_len = max(len(w) for w in words)
        return {w for w in words if len(w) == max_len}

    else:
        raise ValueError("Invalid task")
```

```python
# Word Filter by Criteria
```

```python
def filter_words(sentence: str, criteria: str) -> set[str]:
    """Return a set of words filtered by the given criteria."""

    import re

    # Extract alphabetic words only (ignore punctuation, numbers, symbols)
    words = re.findall(r"[A-Za-z]+", sentence)

    # Helper functions
    def is_palindrome(w):
        w_low = w.lower()
        return w_low == w_low[::-1]

    def is_balanced(w):
        w_low = w.lower()
        vowels = sum(1 for ch in w_low if ch in "aeiou")
        consonants = sum(1 for ch in w_low if ch.isalpha() and ch not in "aeiou")
        return (vowels == consonants) and (vowels + consonants > 0)

    def is_alphabetical(w):
        w_low = w.lower()
        return all(w_low[i] <= w_low[i+1] for i in range(len(w_low)-1))

    def is_alphabetical_palindrome(w):
        if not is_palindrome(w):
            return False
        w_low = w.lower()
        mid = (len(w_low) + 1) // 2  # includes middle for odd length
        half = w_low[:mid]
        return is_alphabetical(half)

    def is_balanced_palindrome(w):
        return is_balanced(w) and is_palindrome(w)

    # Map criteria strings to functions
    criteria_map = {
        "palindromes": is_palindrome,
        "balanced": is_balanced,
        "alphabetical": is_alphabetical,
        "alphabetical_palindromes": is_alphabetical_palindrome,
        "balanced_palindromes": is_balanced_palindrome,
    }

    # If unknown criteria → return empty set
    if criteria not in criteria_map:
        return set()

    check = criteria_map[criteria]

    # Keep words in original casing, but filter case-insensitively
    return {word for word in words if check(word)}
```

```python
def middle_element(lst):
    """
    Returns the middle element of a list.
    If the list length is even, returns a tuple of the two middle elements.
    """
    n = len(lst)
    if n == 0:
        return None

    mid = n // 2

    if n % 2 == 1:
        # Odd length → single middle element
        return lst[mid]
    else:
        # Even length → return the two middle elements
        return (lst[mid - 1], lst[mid])
```

```python
def get_words_by_criteria(words, criteria=None):
    """
    Filters words based on the given criteria.

    Args:
        words (list): List of words.
        criteria (str): The filter criteria.
                        Supported: 'uppercase', 'lowercase', 'starts_with_vowel',
                                   'starts_with_consonant', 'numeric', 'alphanumeric'

    Returns:
        list: Filtered list of words matching the criteria, or None if invalid criteria.
    """

    if criteria is None:
        return words  # no filtering

    criteria = criteria.lower()

    vowels = "aeiouAEIOU"

    if criteria == "uppercase":
        return [w for w in words if w.isupper()]

    elif criteria == "lowercase":
        return [w for w in words if w.islower()]

    elif criteria == "starts_with_vowel":
        return [w for w in words if w and w[0] in vowels]

    elif criteria == "starts_with_consonant":
        return [w for w in words if w and w[0].isalpha() and w[0] not in vowels]

    elif criteria == "numeric":
        return [w for w in words if w.isdigit()]

    elif criteria == "alphanumeric":
        return [w for w in words if w.isalnum()]

    else:
        return None  # invalid criteria
```

```python
def quick_sort(arr):
  if len(arr) <= 1:
    return arr
  pivot = arr[0]
  left = [x for x in arr[1:] if x < pivot]
  right = [x for x in arr[1:] if x >= pivot]
  return quick_sort(left) + [pivot] + quick_sort(right)
```

```python
print(quick_sort([1,8,5,5]))
```

```python
# Dijkstra’s Algorithm
```

```python
import heapq

def dijkstra(graph, start):
    dist = {node: float("inf") for node in graph}
    dist[start] = 0
    pq = [(0, start)]

    while pq:
        cur_dist, node = heapq.heappop(pq)
        if cur_dist > dist[node]:
            continue
        for neigh, weight in graph[node]:
            new_dist = cur_dist + weight
            if new_dist < dist[neigh]:
                dist[neigh] = new_dist
                heapq.heappush(pq, (new_dist, neigh))

    return dist

graph = {
    'A':[('B',1),('C',4)],
    'B':[('C',2),('D',5)],
    'C':[('D',1)],
    'D':[]
}
print(dijkstra(graph, 'A'))

```

```python
# Longest Common Subsequence (LCS)
```

```python
def lcs(a, b):
    dp = [[0]*(len(b)+1) for _ in range(len(a)+1)]
    for i in range(1, len(a)+1):
        for j in range(1, len(b)+1):
            if a[i-1] == b[j-1]:
                dp[i][j] = dp[i-1][j-1] + 1
            else:
                dp[i][j] = max(dp[i-1][j], dp[i][j-1])
    return dp[-1][-1]

print(lcs("ABCBDAB", "BDCABC"))
```
