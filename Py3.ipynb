```python
def is_palindrome(n: int) -> bool:
    """
    Checks if an integer is a palindrome.

    Arguments:
    n : int - the integer to check

    Return:
    bool - True if the integer is a palindrome, False otherwise
    """
    s = str(n)
    return s == s[::-1]
```

```python
def rotate_list(lst: list, k: int) -> list:
    '''
    Given a list of items and an integer k, rotate the list to the right by k steps.
    '''
    n = len(lst)
    k = k % n  # Handle cases when k > n
    return lst[-k:] + lst[:-k]
```

```python
#Count Vowels and Consonants
```

```python
text = input("Enter a string: ").lower()
vowels = "aeiou"
vowel_count = consonant_count = 0

for ch in text:
    if ch.isalpha():
        if ch in vowels:
            vowel_count += 1
        else:
            consonant_count += 1

print("Vowels:", vowel_count)
print("Consonants:", consonant_count)
```

```python
#Reverse a Number
```

```python
num = int(input("Enter a number: "))
rev = 0

while num != 0:
    digit = num % 10
    rev = rev * 10 + digit
    num //= 10

print("Reversed number:", rev)
```

```python
num = int(input("Enter a number:"))
rev  = 0

while num != 0:
  digit  = num % 10
  rev  = rev * 10 + digit
  num //= 10

  print("Reverse number:", rev)
```

```python
#This program takes an integer input from the user and reverses it using a while loop. It extracts the last digit of the number using the modulus operator (`% 10`), adds it to the reversed number by multiplying the existing reversed value by 10 and adding the digit, and then removes the last digit from the original number using integer division (`// 10`). This process repeats until the original number becomes zero, and finally, the reversed number is printed. For example, if the user enters 1234, the output will be 4321.
```
